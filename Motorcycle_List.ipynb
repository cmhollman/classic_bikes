```python
import pandas as pd
```

```python
#importing data, dropping unwanted columns
df = pd.read_csv('Data/list_curated.csv', low_memory=False)
df = df[['Brand', 'Model', 'Year', 'Category', 'Rating', 'Engine stroke']]
df.head()
```

```python
df['Category'].unique()
```

```python
#specifying only majoy categories
cat_list = ['Classic', 'Sport', 'Touring', 'Allround', 'Sport touring', 'Enduro / offroad']
df = df[df['Year'].between(1960,1980)]
df = df[df['Category'].isin(cat_list)]
df.head()
```

```python
#dropping bikes without ratings, 
df.dropna(inplace=True)
```

```python
#sorting by Rating, Specific for Classic category
rate_df = df.sort_values('Rating', ascending = False)
classics_rated = rate_df[rate_df['Category'] == 'Classic']
sport_rated = rate_df[rate_df['Category'] == 'Sport']
touring_rated = rate_df[rate_df['Category'] == 'Touring']
allround_rated = rate_df[rate_df['Category'] == 'Allround']
sport_tour_rated = rate_df[rate_df['Category'] == 'Sport touring']
enduro_rated = rate_df[rate_df['Category'] == 'Enduro / offroad']
```
